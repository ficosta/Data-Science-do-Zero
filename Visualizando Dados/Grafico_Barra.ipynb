```python
import matplotlib.pyplot as plt
```

```python
movies = ["Annie Hall", "Been-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
#Exemplo livro

# Barras possuem o tamanho padrao de 0.8 entao adicionaremos 0.1
# as coordenadas a esquerda para que cada barra seja centralizada
xs = [i + 0.1 for i, _ in enumerate(movies)]

#cria um gr치fico de linhas, anos no eixo x, gdp no eixo y
plt.bar(xs, num_oscars)

#adiciona um titulo
plt.title('Meus filmes favoritos')

#Adiciona um selo no eixo y
plt.ylabel('# de premiacoes')

plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)
plt.show()
```
